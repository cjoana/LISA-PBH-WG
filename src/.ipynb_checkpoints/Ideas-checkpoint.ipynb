```python
import numpy as np
import os
import matplotlib as mp
import matplotlib.pyplot as plt
```

# Bouncing to inflati1on

```python
Mp = 1/np.sqrt(8*np.pi)
V = 1e-3
k =  V/10
# H2 = V  /(3*Mp**2)   - k/a**2


t=1
a = 1
adot = - a * np.sqrt( V  /(3*Mp**2)   - k/a**2 )
T = (-2 * V)
adotdot = -a *(4*np.pi* T)/3

Hcons = (adot/a)**2 -  V  /(3*Mp**2) + k/a**2

aout = [a]
adout = [adot]
addout = [adotdot]
tout = [t]
Cout = [Hcons]



tend = 15  *0.5* np.abs(V)**-0.5
dt = 1e-4

while t < tend:
    
    a = a + adot * dt
    adot = adot + adotdot*dt
    T = (-2 * V)
    adotdot = - a *(4*np.pi* T)/3
    
    Hcons = (adot/a)**2 -  V  /(3*Mp**2) + k/a**2

    
    t = t+dt
    aout.append(a)
    adout.append(adot)
    addout.append(adotdot)
    tout.append(t)
    Cout.append(Hcons)
    

plt.plot(tout,aout, label=r"$a$")
plt.plot(tout,adout, label=r"$\dot a$")
plt.plot(tout,addout, label=r"$\ddot a$")
plt.plot(tout,np.abs(Cout), label=r"$C$")
# plt.xscale("log")
plt.yscale("log")
# plt.ylim(1e-3, 10)
plt.legend()
plt.show()
```
